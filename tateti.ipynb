```python
from IPython.display import clear_output
import random
def tablero(board):
    clear_output()
    print(' '+board[7]+'  |  '+''+board[8]+'  |  '+' '+board[9])
    print('\n'+'-----------------'+ '\n')
    print(' '+board[4]+'  |  '+''+board[5]+'  |  '+' '+board[6])
    print('\n'+'-----------------'+ '\n')
    print(' '+board[1]+'  |  '+''+board[2]+'  |  '+' '+board[3])
board=['#','X','O','X','O','X','O','X','O','X']
def player_input():
  
    marker=''
    while marker!='X' and marker!='O':
        marker=input('Player 1 quiere "X" o "O"').upper()
    if marker=='X':
            player1='X'
            player2='O'
    else:
            player1='O'
            player2='X'
    return(player1,player2)
def lugarmarca(board,marker,posicion):
    board[posicion]=marker
def win_check(board,mark):
    
    return ((board[7] == mark and board[8] == mark and board[9] == mark) or # across the top
    (board[4] == mark and board[5] == mark and board[6] == mark) or # across the middle
    (board[1] == mark and board[2] == mark and board[3] == mark) or # across the bottom
    (board[7] == mark and board[4] == mark and board[1] == mark) or # down the middle
    (board[8] == mark and board[5] == mark and board[2] == mark) or # down the middle
    (board[9] == mark and board[6] == mark and board[3] == mark) or # down the right side
    (board[7] == mark and board[5] == mark and board[3] == mark) or # diagonal
    (board[9] == mark and board[5] == mark and board[1] == mark)) # diagonal

def space_check(board, position):
    return board[position] == ''
def full(board):
     if '' in board:
            return False


def choose_first():
    if random.randint(0, 1) == 0:
        return 'Player 2'
    else:
        return 'Player 1'
def desicionplayer(board):
    numelejido=0
    while numelejido not in [1,2,3,4,5,6,7,8,9] or not space_check(board,numelejido):
        numelejido=int(input('eleji un numero del 1-9'))
    return numelejido
def jugardenuevo():
    return input('¿Desea jugar de vuelta?').startswith('s')

######################main 
print('Bienvenido al jogo manito')

while True:
    board=['#']+['']*9
    marker1,marker2=player_input()
    empezar=input('¿Desea empezar a jugar?').lower()
    if empezar.startswith('s'):
        empezar=True
        
    else:
        empezar=False
    primero=choose_first()
    print('el jugador que comienza es'+primero) 
    while empezar==True:
       
       
        if primero=='Player 1':
           
            tablero(board)
            posicion1=desicionplayer(board)
            lugarmarca(board,marker1,posicion1)
            
            if full(board):
                tablero(board)
                print('Hubo empate')
                empezar=False
            else: 
                if win_check(board,marker1):
                    tablero(board)
                    print('Felicitaciones Ganaste')            
                    empezar=False
                else:
                    primero='Player 2'
        elif primero=='Player 2':
            print('tu turno player 2')
            tablero(board)
            posicion=desicionplayer(board)
            lugarmarca(board,marker2,posicion)
            
            if full(board):
                tablero(board)
                print('Hubo empate')
                empezar=False
            else:
                if win_check(board,marker2):
                    tablero(board)
                    print('Felicitaciones Ganaste')
                    empezar=False
                else:
                    primero='Player 1'
    if not jugardenuevo():
        break
        
            
                
        
        
            
            
        
    
            
    
    
    
    
        
```
